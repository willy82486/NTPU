```python
import numpy as np
import os
from os import listdir 
from os.path import isfile, join
import cv2
import imutils

import tensorflow as tf
import tensorflow.keras as keras

label_folder = []
total_size = 0
data_path = r"C:\Users\jason\Videos\Training data"

#os.walk() generates the file names(dirpath, dirnames, filenames) 
#in a directory tree by walking the tree either top-down or bottom-up.
for root, dirts, files in os.walk(data_path): 
    for dirt in dirts:
        label_folder.append(dirt)
    total_size += len(files)

    
print("found",total_size,"files.")
print("folder:",label_folder)
```

```python
base_x_train = []
base_y_train = []

for i in range(len(label_folder)):
    labelPath = data_path+r'\\'+label_folder[i]
    
    #listdir() returns a list containing the names of the entries in the directory given by path.
    #isfile() is used to check whether the specified path is an existing regular file or not.
    FileName = [f for f in listdir(labelPath) if isfile(join(labelPath, f))]
    
    for j in range(len(FileName)):
        path = labelPath+r'\\'+FileName[j]
        
        #use cv2.imread read image.
        img = cv2.imread(path,cv2.IMREAD_COLOR)
        
        base_x_train.append(img)
        base_y_train.append(label_folder[i])
        rotated = imutils.rotate(img, 15)
        base_x_train.append(rotated)
        base_y_train.append(label_folder[i])
        rotated = imutils.rotate(img, -15)
        base_x_train.append(rotated)
        base_y_train.append(label_folder[i])
        base_x_train.append(cv2.flip(img, 1))
        base_y_train.append(label_folder[i])


print(np.array(base_x_train).shape)
print(np.array(base_y_train).shape)
```

```python
total_size2 = 0
data_path2 = r"C:\Users\jason\Videos\410785036"
    
base_x_test = []

for root, dirts, files in os.walk(data_path2):
    total_size2 += len(files)
print(total_size2)

FileName2 = [f for f in listdir(data_path2) if isfile(join(data_path2, f))]
#print(FileName)
    
for j, fname in enumerate(FileName2):
    path2 = data_path2 + r"\\" + fname
    img = cv2.imread(path2, cv2.IMREAD_COLOR)        
    base_x_test.append(img)

x_test = np.array(base_x_test)
print(x_test.shape)
```

```python
from tensorflow.keras.utils import to_categorical

#Convert a category vector to a binary (0 or 1) matrix-type representation

base_y_train = to_categorical(base_y_train)


print(np.array(base_x_train).shape)
print(np.array(base_y_train).shape)
```

```python
base_x_train = np.array(base_x_train).astype('float32')
x_test = x_test.astype('float32')
base_x_train /= 255
x_test /= 255
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(64, (3, 3), padding="same", kernel_initializer="he_normal", activation='relu', input_shape=(32, 32, 3)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(rate=0.3),
    tf.keras.layers.Conv2D(64, (3, 3), padding="same", kernel_initializer="he_normal", activation='relu', input_shape=(32, 32, 3)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(rate=0.3),
    tf.keras.layers.MaxPool2D(pool_size=2),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(rate=0.3),
    tf.keras.layers.Conv2D(128, (3, 3), padding="same", kernel_initializer="he_normal", activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(rate=0.3),
    tf.keras.layers.Conv2D(128, (3, 3), padding="same", kernel_initializer="he_normal", activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(rate=0.3),
    tf.keras.layers.MaxPool2D(pool_size=2),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(rate=0.3),
    tf.keras.layers.Conv2D(256, (3, 3), padding="same", kernel_initializer="he_normal", activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(rate=0.3),
    tf.keras.layers.Conv2D(256, (3, 3), padding="same", kernel_initializer="he_normal", activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(rate=0.3),
    tf.keras.layers.MaxPool2D(pool_size=2),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(rate=0.3),
    tf.keras.layers.Conv2D(512, (3, 3), padding="same", kernel_initializer="he_normal", activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(rate=0.3),
    tf.keras.layers.Conv2D(512, (3, 3), padding="same", kernel_initializer="he_normal", activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(rate=0.3),
    tf.keras.layers.MaxPool2D(pool_size=2),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(rate=0.3),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, kernel_initializer="he_normal", activation="relu"),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(256, kernel_initializer="he_normal", activation="relu"),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(128, kernel_initializer="he_normal", activation="relu"),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(64, kernel_initializer="he_normal", activation="relu"),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(rate=0.2),
    tf.keras.layers.Dense(10, activation="softmax")
])
```

```python
print(model.summary())
```

```python
epoch = 40
batch_size = 64

model.compile(
    loss="categorical_crossentropy", 
    optimizer="adam", 
    metrics=["accuracy"]
)

p = np.random.permutation(base_x_train.shape[0])
base_x_train = base_x_train[p]
base_y_train = base_y_train[p]

train_num = 20000
x_train, x_valid = np.array(base_x_train)[:train_num], np.array(base_x_train)[train_num:]
y_train, y_valid = np.array(base_y_train)[:train_num], np.array(base_y_train)[train_num:]
history = model.fit(
    x_train, 
    y_train, 
    epochs=epoch, 
    batch_size=batch_size,
    validation_data=(x_valid, y_valid)
)
```
