```python
import numpy as np
from tensorflow import keras
import os
import cv2
from os import listdir
from os.path import isfile
from os.path import join
class NN:
    def __init__(self, sizes, lr = 0.05):
        self.sizes = sizes
        self.lr = lr
        self.paras = self.iniparas()
        
    def iniparas(self):   
        input_layer = self.sizes[0]
        hidden_layer_1 = self.sizes[1]
        hidden_layer_2 = self.sizes[2]
        output_layer = self.sizes[3]

        paras = {
            'w1':np.random.normal(0., 1.,(hidden_layer_1, input_layer)) * np.sqrt(1. / hidden_layer_1),
            'w2':np.random.normal(0., 1.,(hidden_layer_2, hidden_layer_1)) * np.sqrt(1. / hidden_layer_2),
            'w3':np.random.normal(0., 1.,(output_layer, hidden_layer_2)) * np.sqrt(1. / output_layer),
        }

        return paras
    
    def forward_pass(self, x):
        paras = self.paras       
        paras['a0'] = x

        # hidden layer 1
        paras['z1'] = np.dot(paras["w1"], paras['a0'])
        paras['a1'] = self.Sigmoid(paras['z1'])

        # hidden layer 2
        paras['z2'] = np.dot(paras["w2"], paras['a1'])
        paras['a2'] = self.Sigmoid(paras['z2'])

        # output layer
        paras['z3'] = np.dot(paras["w3"], paras['a2'])
        paras['a3'] = self.Softmax(paras['z3'])

        return paras['a3']
    
    def backward_pass(self, y, output):
        paras = self.paras
        changed_w = {}
        
        # update w
        error = 2 * (output - y) / output.shape[0] * self.Softmax(paras['z3'], derivative = True)
        changed_w['w3'] = np.outer(error, paras['a2'])

        error = np.dot(paras['w3'].T, error) * self.Sigmoid(paras['z2'], derivative = True)
        changed_w['w2'] = np.outer(error, paras['a1'])
        
        error = np.dot(paras['w2'].T, error) * self.Sigmoid(paras['z1'], derivative = True)
        changed_w['w1'] = np.outer(error, paras['a0'])

        return changed_w
    
    def Sigmoid(self, z, derivative = False):
        if derivative:
            return (np.exp(-z))/((np.exp(-z)+1)**2)
        return 1/(1 + np.exp(-z))
    
    def Softmax(self, z, derivative = False):
        exps = np.exp(z - z.max())
        if derivative:
            return exps / np.sum(exps, axis=0) * (1 - exps / np.sum(exps, axis=0))
        return exps / np.sum(exps, axis=0)
    
    def update(self, changed_w):
        for key, value in changed_w.items():
            delW = self.lr * value
            self.paras[key] -= delW
        
    def train(self, x_train, y_train):
        for x, y in zip(x_train, y_train):
            x = x.reshape(784, 1)
            y = y.reshape(8, 1)
            f_result = self.forward_pass(x)
            b_result = self.backward_pass(y, f_result)
            self.update(b_result)
        self.lr *= 0.95
            
    def validate(self, x_test, y_test):
        error = 0
        for x, y in zip(x_test, y_test):
            x = x.reshape(784, 1)
            y = y.reshape(8, 1)
            f_result = self.forward_pass(x)
            if np.argmax(f_result) != np.argmax(y):
                error += 1
        print(1 - error/ len(x_test))
        
    def predict(self, real_x_test):
        filecounter = 0
        with open("410785036.txt", 'w') as f:
            for x in real_x_test:
                x = x.reshape(784, 1)
                f_result = self.forward_pass(x)
                path = data_path + r'\\' + FileName[filecounter]
                txt_name = os.path.splitext(FileName[filecounter].split('.')[0])[0]
                f.write(txt_name + " " + (np.argmax(f_result)).astype(str) + "\n")
                filecounter += 1
```
